```python
#Downloads a FITS file from the STScI DeepMerge dataset, caches it locally, and opens it for reading with Astropy's FITS tools
# https://archive.stsci.edu/hlsps/deepmerge/

from astropy.io import fits
from astropy.utils.data import download_file

#Change the pristine version to noicy is needed.(noisy is the only other version)
version = 'pristine'

file_url = 'https://archive.stsci.edu/hlsps/deepmerge/hlsp_deepmerge_hst-jwst_acs-wfc3-nircam_illustris-z2_f814w-f160w-f356w_v1_sim-'+version+'.fits'

#this does 2 things:
#1. download the FITS file locally and the show_progress is displayed. Waits upto 120 secnds before the error is shown
#2. opens the FITS file in HDUList format
hdu = fits.open(download_file(file_url, cache=True, show_progress=True, timeout=120)) 
```

```python
# Each header contains info about the data it describes, like a label or description
# hdu[0].header accesses the metadata of the main data part of the FITS file
# Known as the Primary HDU. This often contains the main image or primary data and header about the overall file
print("The main data part of the FITS file contains the following information:")
hdu[0].header
```

Indicates that the FITS file may contain extensions beyond the primary data.  
<strong>NAME1, NAME2, NAME3, NAME4:</strong>  
Names describing the role of each axis:

```python
# Print the headers where you may find clues about the layers (e.g., band names)
# Show the first galaxy image in each of the 3 layers to visually compare what they represent

# Access the primary header
header = hdu[0].header
# Extract the comment of the 'NAME3' keyword which contains filter names
filters_comment = header.comments['NAME3']  # This should be: 'F814W,F356W,F160W'
# Split the filter names into a list
filters_list = filters_comment.split(',')

print(filters_list)

# Visualize the first image's 3 layers:

fig, axs = plt.subplots(1, 3, figsize=(12, 4))
for i in range(3):
    layer_image = hdu[0].data[0, i, :, :]  # first galaxy, ith layer
    axs[i].imshow(layer_image, cmap='gray')
    axs[i].set_title(f'Layer {i}')
    axs[i].axis('off')
plt.show()
```

<small>

These three layers are images of the same galaxies taken through different "color" filters in infrared and optical wavelengths. Each filter highlights different physical properties or components of the galaxies liketheir structures, star formation, and dust content more comprehensively

**F814W:** A filter on Hubble that sees light just beyond red (near‑infrared, 814 nm). Good for showing galaxy shapes and details.

**F356W:** A filter on JWST that sees mid‑infrared light (3.56 μm). Reveals cooler things like dust, older stars, and hidden star‑forming areas.

**F160W:** A filter on Hubble that sees near‑infrared light (1.6 μm). Helps see through dust to detect stars and hidden galaxy features.
